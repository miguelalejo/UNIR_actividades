### 1. Crea una función en Python que calcule el máximo común divisor de dos números.

```python
def calcular_maximo_comun_divisor(numeroA,numeroB):
    if numeroA == 0 and numeroB == 0 :
         raise Exception('El máximo común divisor no está definido para a=b=0.')
    if numeroA < 0 or numeroB < 0 :
        raise Exception('a y b deben ser enteros positivos')
    max_rango=numeroA if numeroA>numeroB else numeroB
    max_divisor=1
    for divisor in range(1,max_rango+1):        
        if numeroA%divisor==0 and numeroB%divisor==0:
            max_divisor = divisor
    return max_divisor 
print("El máximo común divisor de 12 y 18 es %d"%calcular_maximo_comun_divisor(12, 18))
```

### 2. Crea una función en Python que calcule el mínimo común múltiplo de dos números.

>  Matemáticamente  se cumple  **a·b = M.C.D.(a,b)·M.C.M.(a,b)** 

```python
def calcular_minimo_comun_multiplo(numeroA,numeroB):    
    min_multiplo=(numeroA*numeroB)/calcular_maximo_comun_divisor(numeroA,numeroB)
    return min_multiplo
print("El mínimo común múltiplo de 16 y 117 es %d"%calcular_minimo_comun_multiplo(16, 117))
```

# Segunda parte: Procesado y limpieza de datos con funciones de cadena

### 1. Lee el fichero proporcionado.

```python
def leer_archivo():
    archivo = open("FicheroALeer.txt")
    numero_lineas = 0
    lista_texto = archivo.readlines()
    texto=""
    for linea_texto in lista_texto:
        texto = texto + linea_texto
    numero_lineas = len(lista_texto)
    archivo.close()
    return texto,numero_lineas
texto, numero_lineas = leer_archivo()
print(texto)
```

### 2. Cuenta el número de líneas del fichero.

```python
print('Número lineas fichero: %s'%numero_lineas)
```

### 3. Cuenta el número de veces que aparece el artículo «el» en el fichero (nota: elemento contiene los caracteres «el» pero no debe contar puesto que no contiene a el artículo «el».

```python
def contar_veces_articulo_el(texto):
    articulo="el"
    lista_palabras = texto.split()
    numero_veces_articulo = 0
    for palabra in lista_palabras:
        if articulo == palabra.lower():
            numero_veces_articulo = numero_veces_articulo+1
    return numero_veces_articulo
numero_veces_articulo = contar_veces_articulo_el(texto)
print('Número veces articulo: %s'%numero_veces_articulo)     
```

### 4.Elimina las tildes del fichero y convierta todas las letras a minúscula.

```python
def eliminar_tildes_convertir_minuscula(texto):
    texto_temp=texto.lower()
    texto_temp=texto_temp.replace("á","a")
    texto_temp=texto_temp.replace("é","e")
    texto_temp=texto_temp.replace("í","i")
    texto_temp=texto_temp.replace("ó","o")
    texto_temp=texto_temp.replace("ú","u")
    return texto_temp
texto_sin_tildes_minuscula = eliminar_tildes_convertir_minuscula(texto) 
print(texto_sin_tildes_minuscula)
```

### 5. Crea un fichero de texto con el nombre «resultado.txt» que contendrá en la primera línea el número de líneas del fichero, en la segunda línea el número de veces que aparece el artículo «el», y a partir de la tercera línea el texto original todo en minúsculas y sin tilde.

```python
archivo_resultado = open("resultado.txt","w+")
archivo_resultado.write("Número de líneas: %d\r" % numero_lineas)
archivo_resultado.write("Número de veces que aparece el artículo «el»: %d\r" % numero_veces_articulo)
archivo_resultado.write(texto_sin_tildes_minuscula)
archivo_resultado.close()
```

# Tercera parte: Procesado de datos con Pandas y Numpy

### Carga en el entorno el «poblacionMunicipios.csv» proporcionado por el profesor respetando las tildes y caracteres españoles.

```python
import pandas as pd
df_provincias_municipios=pd.read_csv('poblacionMunicipios.csv',encoding='utf-8')
display(df_provincias_municipios.head(9))
```

### Elimina las filas correspondientes a municipios sin población

```python
df_provincias_municipios=df_provincias_municipios[df_provincias_municipios['Poblacion']>0]
display(df_provincias_municipios)
```

### Imprime el total de la población de todos los municipios

```python
total_poblacion_municipios = df_provincias_municipios['Poblacion'].sum()
print('Total población: %s'%total_poblacion_municipios)
```

###  Crea un DataFrame de Pandas donde la primera columna sean todas las provincias que aparecen en el archivo, la segunda columna el número total de habitantes por provincia, la tercera la desviación típica en el número total de habitantes por provincia y la cuarta el número total de municipios por provincia.

```python
df_totales_provincia=df_provincias_municipios.groupby('Provincia',as_index=False).agg({'Poblacion': ['sum', 'std'],'CodMunicipio': ['count']})
df_totales_provincia.columns = ['Provincia','TotalPoblacion','DesviacionTipica','NumeroMunicipios']
display(df_totales_provincia)
```

### Carga el archivo «CP_Municipios.csv».

```python
df_municipios=pd.read_csv("CP_Municipios.csv",encoding='utf-8')
display(df_municipios.head(9))
```

### Empleando el código de provincia y el código de municipio, cruza los dos datasets comentados hasta el momento. El resultado final debe ser un DataFrame.

```python
df_cruce = pd.merge(df_provincias_municipios,df_municipios,on=['CodProvincia','CodMunicipio'],how="inner")
display(df_cruce)
```

### Genera un CSV llamado «faltan.csv» con todos los códigos postales que estén en «CP_Municipios.csv» pero que no tengan en «poblacionMunicipios.csv» ningún municipio con el que poder cruzar. Si todo cruza correctamente el archivo «faltan.csv» estará vacío. Se puede decir que «faltan.csv» son los códigos postales que se han perdido en el cruce a la hora de considerar la población. 

```python
df_faltan_codigos_postales= df_municipios[~(df_municipios['CodMunicipio'].isin(df_provincias_municipios['CodMunicipio']))]
print(df_faltan_codigos_postales)
df_faltan_codigos_postales.to_csv('faltan.csv', index=False)
```

<ul>
    <li>Código postal.</li>
    <li>Número de municipios que tienen dicho código postal asignado.</li>
